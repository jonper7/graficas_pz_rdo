```python
import openpyxl
from openpyxl.styles import Border, Side

# Función para aplicar borde derecho medio y mantener los bordes existentes
def aplicar_borde_derecho(sheet, min_row, max_row, col):
    medium_right = Side(border_style="medium", color="000000")  
    for row in range(min_row, max_row +1):
        cell = sheet.cell(row=row, column=col)
        
        # Mantener los bordes existentes (si los hay) y agregar el borde derecho
        current_border = cell.border
        new_border = Border(
            left=current_border.left,
            right=medium_right,  # Cambia el borde derecho a "medium"
            top=current_border.top,
            bottom=current_border.bottom
        )
        cell.border = new_border

# Función para aplicar el borde en varias hojas
def aplicar_borde_en_varias_hojas(ruta_salida, hojas, min_row, max_row, col):
    # Cargar el archivo Excel
    workbook = openpyxl.load_workbook(ruta_salida)

    # Aplicar el borde derecho en cada hoja
    for hoja in hojas:
        sheet = workbook[hoja]
        aplicar_borde_derecho(sheet, min_row, max_row, col)

    # Guardar el archivo Excel con los cambios
    workbook.save(ruta_salida)

# Especificar la ruta y las hojas a modificar

ruta_salida = r"Reporte\2500-DRT-MGP-000-V0.xlsx"
hojas = [

    "PA22-01-T", "PA22-02-T", "PA22-06-T", "PA22-07-T",
    "PA22-09-T", "PA22-10-T", "PA22-12-T", "PC22-03-T",
    "PC22-04-T", "PC22-05-T", "PC22-08-T", "PC22-11-T",
    "PC22-14-T", "PC22-15-T", "PC23-01-T", "PC23-02-T_PC23-02A-T",
    "PC23-04-T", "PC23-05-T", "PC23-06-T", "PC23-07-T",
    "PC23-09-T", "PC23-10-T", "PC23-15A-T_PC23-15B-T", "PC23-16-T",
    "PC23-19-T", "PC23-23-T", "PC23-24-T", "PC23-25-T",
    "PC23-26-T", "PC24-03-T", "PC24-04-T", "PA23-03-T",
    "PA23-08-T", "PA23-11-T", "PA23-12-T", "PA23-13-T",
    "PA23-14-T", "PA23-17A-T", "PA23-18-T", "PA23-20A-T",
    "PA23-21-T", "PA23-22-T", "PA23-27-T", "PA23-28-T",
    "PA23-29-T", "PA23-30-T", "PA24-01-A-T", "PA24-02-A-T",
    "PA24-05-A-T", "PA24-10-A-T", "PA24-28-T", "PA24-FT-128-T",
    "PA24-11-T", "PA24-01-S", "PA24-02-S", "PA24-03-S",
    "PA24-04-S", "PA24-05-S", "PA24-01-T", "PA24-02-T",
    "PA24-03-T", "PA24-04-A-T", "PA24-05-T", "PA24-06-T",
    "PA24-07-T", "PA24-08-T", "PA24-09-T", "PA24-10-T",
    "PA24-12-T", "PA24-03-A-T"

]

# Aplicar el borde derecho medio en todas las hojas especificadas
aplicar_borde_en_varias_hojas(ruta_salida, hojas, min_row=2, max_row=61, col=45)

print('Bordes aplicados con éxito')
```
